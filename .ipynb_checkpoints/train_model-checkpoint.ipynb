```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import xgboost as xgb

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
import seaborn as sns 
sns.set()
```

Notebook for training models, creating dataset for Multi Level Model, some data viz

```python
def time_remaining(period,minutes,seconds):
    mins = 0
    if period == 1:
        mins = 36
    elif period == 2:
        mins = 24
    elif period == 3:
        mins = 12
    mins = mins + minutes
    secs = mins * 60
    secs = secs + seconds
    return secs
```

```python
#theres like 40 different action types, group them together
def shot_type(shot_name):
    shot_name = shot_name.lower()
    if "dunk" in shot_name:
        return "Dunk"
    elif "layup" in shot_name:
        return "Layup"
    elif "jump shot" in shot_name:
        return "Jump Shot"
    elif "hook" in shot_name:
        return "Hook"
    else:
        return "Other"
```

```python
# get shot cluster for each shot in the dataset
def shot_zone(distance):
    if distance < 5:
        return 1
    elif distance >= 5 and distance < 10:
        return 2
    elif distance >= 10 and distance < 20:
        return 3
    else:
        return 4

def shot_cluster(zone,short,close,mid,long):
    if zone == 1:
        return short
    elif zone == 2:
        return close
    elif zone == 3:
        return mid
    elif zone == 4:
        return long
```

```python
#load all shots with player cluster info
all_shots = pd.read_csv("./data/all_shots_players.csv")

#sort by game date so we can do some forcasting once we have the model
all_shots.sort_values(by=['GAME_DATE'],ascending=True,inplace=True)
all_shots = all_shots.reset_index()

#removing team id, game dates, etc.
feature_label_cols = all_shots[['PLAYER_ID','ACTION_TYPE','SHOT_ZONE_BASIC','SHOT_ZONE_AREA','LOC_X','LOC_Y','1st_closest_defender_dist', '1st_closest_defender_angle',
       'period','minutes_remaining','seconds_remaining','2nd_closest_defender_dist', '2nd_closest_defender_angle','x','y','touch_time','defender_distance','shot_distance',
       '3rd_closest_defender_dist', '3rd_closest_defender_angle','shot_clock','short_cluster','close_cluster','mid_cluster','long_cluster','dribbles','SHOT_MADE_FLAG','shooter_x','shooter_y']]
#grouping the massive amount of different shot descriptions
feature_label_cols['shot_type'] = feature_label_cols['ACTION_TYPE'].apply(lambda x: shot_type(x))

#one hot encodings for shot types, zones
one_hot_zone = pd.get_dummies(feature_label_cols['SHOT_ZONE_BASIC'])
one_hot_shot_type = pd.get_dummies(feature_label_cols['shot_type'])
one_hot_shot_area = pd.get_dummies(feature_label_cols['SHOT_ZONE_AREA'])
#switch out categorical variables for numeric
flc = feature_label_cols.join([one_hot_zone,one_hot_shot_type,one_hot_shot_area])
#training size
train_num = int(len(flc) * 0.75)

#get distance zone for shot so we can use the appropriate shooting cluster for the player's shooting ability
flc['shot_zone'] = flc['shot_distance'].apply(lambda x: shot_zone(x))
flc['shot_cluster'] = flc.apply(lambda x: shot_cluster(x.shot_zone,x.short_cluster,x.close_cluster,x.mid_cluster,x.long_cluster),axis = 1)
flc['time_remaining'] = flc.apply(lambda x: time_remaining(x['period'],x['minutes_remaining'],x['seconds_remaining']),axis = 1)
```

```python
#save this to csv so I can load it later
flc.to_csv("./data/features_and_labels.csv")
```

```python
#just get location, game data for first model
first_model_ = flc[['x', 'y', 'time_remaining','shot_clock','Above the Break 3', 'Backcourt', 'In The Paint (Non-RA)',
       'Left Corner 3', 'Mid-Range', 'Restricted Area', 'Right Corner 3','shot_distance',
       'Dunk', 'Hook', 'Jump Shot', 'Layup', 'Other','SHOT_MADE_FLAG']]
#location + pressure + dribbling data for second model
second_model_ = flc[['1st_closest_defender_dist',
       '1st_closest_defender_angle', '2nd_closest_defender_dist',
       '2nd_closest_defender_angle','3rd_closest_defender_dist','3rd_closest_defender_angle',
       'x', 'y', 'touch_time', 'defender_distance','time_remaining',
       'shot_distance','shot_clock','dribbles',
       'Above the Break 3', 'Backcourt', 'In The Paint (Non-RA)',
       'Left Corner 3', 'Mid-Range', 'Restricted Area', 'Right Corner 3',
       'Dunk', 'Hook', 'Jump Shot', 'Layup', 'Other','shooter_x','shooter_y','SHOT_MADE_FLAG']]
# use second model + player info for third multilevel model
#or should this just be each output fomr second model + player info?
```

```python
#first we have rows in chronological order to seperate train & test. This is b/c we want to test whole games in the future
def train_test(model,train_num):
    train_set = model.iloc[:train_num,:]
    test_set = model.iloc[train_num:,:]
    #drop na rows - found ~600 in total out of 97497
    train_set = train_set.dropna()
    test_set = test_set.dropna()
    #now we randomize rows
    train_set = train_set.sample(frac = 1)
    test_set = test_set.sample(frac = 1)
    #get X, y
    y_train = train_set['SHOT_MADE_FLAG'].values
    y_test = test_set['SHOT_MADE_FLAG'].values
    X_train = train_set.drop("SHOT_MADE_FLAG",axis = 1).values
    X_test = test_set.drop("SHOT_MADE_FLAG",axis = 1).values
    
    return X_train, X_test, y_train, y_test
```

```python
#grid search - first model
first_model_.dropna(inplace=True)
gbc= GradientBoostingClassifier()
X_train,X_test,y_train,y_test = train_test(first_model_,int(len(first_model_) * 0.75))
grid_search = GridSearchCV(gbc,{'loss':['deviance'],'n_estimators':[100,150,200,250],'max_depth':[2,3,4,5]},cv=5)
grid_search.fit(X_train,y_train)
print(grid_search.best_score_)
#grid search - second model
second_model_.dropna(inplace=True)
gbc= GradientBoostingClassifier()
X_train,X_test,y_train,y_test = train_test(second_model_,int(len(second_model_) * 0.75))
grid_search_sec = GridSearchCV(gbc,{'loss':['deviance'],'n_estimators':[100,150,200,250],'max_depth':[2,3,4,5]},cv=5)
X_train.shape
grid_search_sec.fit(X_train,y_train)
print(grid_search_sec.best_score_)
```

```python
def model_scores(X_train, X_test, y_train, y_test):
    '''
    get scores from different models
    '''
    clf = LogisticRegression(max_iter = 200)
    clf.fit(X_train,y_train)
    logr_score = clf.score(X_test,y_test)
    
    gb_clf = GradientBoostingClassifier()
    gb_clf.fit(X_train,y_train)
    gbc_score = gb_clf.score(X_test, y_test)
    
    clf_f = RandomForestClassifier()
    clf_f.fit(X_train,y_train)
    rfc_score = clf.score(X_test, y_test)
    
    return logr_score,gbc_score,rfc_score
```

```python
def models_(X_train, X_test, y_train, y_test):
    '''
    Return the models for further analysis
    '''
    #get scores from different model
    clf = LogisticRegression(max_iter = 200)
    clf.fit(X_train,y_train)
    logr_score = clf.score(X_test,y_test)
    print("Logistic: ",logr_score)
    gb_clf = GradientBoostingClassifier()
    gb_clf.fit(X_train,y_train)
    gbc_score = gb_clf.score(X_test, y_test)
    print("GB: ",gbc_score)
    clf_f = RandomForestClassifier()
    clf_f.fit(X_train,y_train)
    rfc_score = clf.score(X_test, y_test)
    print("RF: ",rfc_score)
    return clf,gb_clf,clf_f
```

```python
#get train, test data
X_train,X_test,y_train,y_test = train_test(first_model_,int(len(first_model_) * 0.75))
#get model scores
logr_score_first,gbc_score_first,rfc_score_first = model_scores(X_train,X_test,y_train,y_test)
#print model scores
first_model_scores = [logr_score_first,gbc_score_first,rfc_score_first,xgb_score_first]
print("FIRST MODEL SCORES: ",first_model_scores)

#same for second model
X_train,X_test,y_train,y_test = train_test(second_model_,int(len(second_model_) * 0.75))
logr_score_first,gbc_score_first,frc_score_first = model_scores(X_train,X_test,y_train,y_test)
second_model_scores = [logr_score_first,gbc_score_first,rfc_score_first,xgb_score_first]
print("SECOND MODEL SCORES: ",second_model_scores)
```

```python
#determine if shot is corner 3 above the break 3 or not a 3
def three_type(abv,lc,rc):
    if abv == 1:
        return "above_break"
    elif lc == 1 or rc == 1:
        return "corner"
    else:
        return "two"
```

```python
first_model_output = multi_level_model_first.drop(['shot_zone','shot_cluster'],axis = 1)
X_train,X_test,y_train,y_test = train_test(first_model_output,int(len(first_model_output) * 0.75))
#get the best model from our grid search to predict values
train_logs = grid_search.predict_log_proba(X_train)
test_logs = grid_search.predict_log_proba(X_test)
logs_probs = np.concatenate((train_logs,test_logs))[:,1]
#drop na so we have same amount of rows
multi_level_model_first.dropna(inplace=True)
multi_level_model_first['first_model_value'] = logs_probs * 100
multi_level_model_first[['first_model_value','shot_zone','shot_cluster','SHOT_MADE_FLAG']].to_csv("./data/multi_level_model_first.csv")

#doing the same for the second model
sec_model_output = multi_level_model_second.drop(['shot_zone','shot_cluster'],axis = 1)
X_train,X_test,y_train,y_test = train_test(sec_model_output,int(len(sec_model_output) * 0.75))
#get the best model from our grid search to predict values
train_logs = grid_search_sec.predict_log_proba(X_train)
test_logs = grid_search_sec.predict_log_proba(X_test)
logs_probs = np.concatenate((train_logs,test_logs))[:,1]
#drop na so we have same amount of rows
multi_level_model_second.dropna(inplace=True)
multi_level_model_second['sec_model_value'] = logs_probs * 100
multi_level_model_second[['sec_model_value','shot_zone','shot_cluster','SHOT_MADE_FLAG']].to_csv("./data/multi_level_model_second.csv")
```

```python
#get roc data to graph roc curve, compute auroc
def roc_curve_and_score(model,x_test,y_test):
    prediction_probs = model.predict_proba(x_test)[:,1]
    false_positive_rate, true_positive_rate, _ = roc_curve(y_test, prediction_probs)
    auroc = roc_auc_score(y_test,prediction_probs)
    return false_positive_rate,true_positive_rate,auroc

X_train_first, X_test_first, y_train_first, y_test_first = train_test(first_model_,int(len(first_model_) * 0.75))
lr_clf,gb_clf,rf_clf = models_(X_train_first, X_test_first, y_train_first, y_test_first)
X_train_second, X_test_second, y_train_second, y_test_second = train_test(second_model_,int(len(second_model_) * 0.75))
lr_clf_sec,gb_clf_sec,rf_clf_sec = models_(X_train_second, X_test_second, y_train_second, y_test_second)
```

```python
#ROC CURVES
gb_fpr,gb_tpr,gb_auroc = roc_curve_and_score(gb_clf,X_test_first,y_test_first)
rf_fpr,rf_tpr,rf_auroc = roc_curve_and_score(rf_clf,X_test_first,y_test_first)
#xgb_fpr,xgb_tpr,xgb_auroc = roc_curve_and_score(xgb_clf,X_test,y_test)
lr_fpr,lr_tpr,lr_auroc = roc_curve_and_score(lr_clf,X_test_first,y_test_first)

plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(gb_fpr, gb_tpr, label = 'Gradient Boosting, AUROC = {}'.format(gb_auroc.round(3)))
plt.plot(rf_fpr, rf_tpr, label = 'Random Forest Classifier, AUROC = {}'.format(rf_auroc.round(3)))
plt.plot(lr_fpr, lr_tpr, label = 'Logistic Regression, AUROC = {}'.format(lr_auroc.round(3)))

plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC Curves For Public Data Models')
plt.legend(loc = 'best')
plt.show()
```
